#    1.  A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
cutlets = pd.read_csv("Cutlets.csv")
cutlets
```

```python
count = len(cutlets)
count
```

```python
mean_A = cutlets['Unit A'].mean()
mean_A
```

```python
std_A = cutlets['Unit A'].std()
std_A
```

```python
z_score = abs(cutlets['Unit B']-mean_A)/((std_A/count)**0.5)
prob_value = norm.cdf(z_score)
```

```python
cutlets['size difference'] = prob_value > 0.95
cutlets 
```

```python
cutlets['size difference'] = cutlets['size difference'].map({True:'significant size difference',False:'Non significant size difference'})
cutlets
```

# 2.   A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
 #  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.


```python
lab = pd.read_csv('LabTAT.csv')
lab
```

```python
labs = lab['Laboratory 1']
labs = labs.append([lab['Laboratory 2'],lab['Laboratory 3'],lab['Laboratory 4']]).reset_index(drop=True)
labs
```

```python
total_mean = labs.mean()
total_mean
```

```python
std_labs = labs.std()
std_labs
```

```python
z_score = abs(labs-total_mean)/((std_labs/len(labs))**0.5)
prob_value = norm.cdf(z_score)
```

```python
pd.DataFrame(labs)
prob = prob_value > 0.95
lab['conclusion 1'] = prob[:120]
lab['conclusion 2'] = prob[120:240]
lab['conclusion 3'] = prob[240:360]
lab['conclusion 4'] = prob[360:]
lab
```

```python
lab = lab[['Laboratory 1','conclusion 1','Laboratory 2','conclusion 2','Laboratory 3','conclusion 3','Laboratory 4','conclusion 4']]
lab
```

```python
lab['conclusion 1'] = lab['conclusion 1'].map({True:'No significant difference',False:'significant difference'})
lab['conclusion 2'] = lab['conclusion 2'].map({True:'No significant difference',False:'significant difference'})
lab['conclusion 3'] = lab['conclusion 3'].map({True:'No significant difference',False:'significant difference'})
lab['conclusion 4'] = lab['conclusion 4'].map({True:'No significant difference',False:'significant difference'})
lab

```

#     3.  Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.


```python
Buyer_ratio = pd.read_csv('BuyerRatio.csv')
Buyer_ratio
```

```python
Buyer_ratio = Buyer_ratio.transpose()
Buyer_ratio.columns = Buyer_ratio.iloc [0]
Buyer_ratio = Buyer_ratio.drop('Observed Values')
Buyer_ratio
```

```python
Buyer_ratio['Ratio'] = Buyer_ratio['Males']/Buyer_ratio['Females']
Buyer_ratio
```

```python
Buyer_ratio = Buyer_ratio.drop(['Males','Females'],axis=1)
Buyer_ratio
```

```python
mean = Buyer_ratio['Ratio'].mean()
mean
```

```python
std = Buyer_ratio['Ratio'].std()
std
```

```python
count = len(Buyer_ratio)
count
```

```python
prob_value = []
z_score = abs(Buyer_ratio['Ratio'].iloc[0]-mean)/((std/count)**0.5)
prob_value.append(norm.cdf(z_score))
z_score = abs(Buyer_ratio['Ratio'].iloc[1]-mean)/((std/count)**0.5)
prob_value.append(norm.cdf(z_score))
z_score = abs(Buyer_ratio['Ratio'].iloc[2]-mean)/((std/count)**0.5)
prob_value.append(norm.cdf(z_score))
z_score = abs(Buyer_ratio['Ratio'].iloc[3]-mean)/((std/count)**0.5)
prob_value.append(norm.cdf(z_score))
prob_value
```

```python
Buyer_ratio['similarity'] = prob_value 
Buyer_ratio['similarity'] = Buyer_ratio['similarity'] > 0.95
Buyer_ratio
```

# 4.     TeleCall uses 4 centers around the globe to process customer order forms. They audit a certain %  of the customer order forms. Any error in order form renders it defective and has to be reworked before processing.  The manager wants to check whether the defective %  varies by centre. Please analyze the data at 5% significance level and help the manager draw appropriate inferences

```python
customer_order = pd.read_csv('Costomer+OrderForm.csv')
customer_order
```

```python
array = ['Region','Defective Percent'],['Philippenes',customer_order['Phillippines'].value_counts().iloc[1]],['Indonesia',customer_order['Indonesia'].value_counts().iloc[1]],['Malta',customer_order['Malta'].value_counts().iloc[1]],['India',customer_order['India'].value_counts().iloc[1]]
defective_percent = pd.DataFrame(array)
defective_percent.columns = defective_percent.iloc[0]
defective_percent = defective_percent.iloc[1:,:]
defective_percent
```

```python
count = len(defective_percent)
mean = defective_percent['Defective Percent'].mean()
std = defective_percent['Defective Percent'].std()
z_score = abs(defective_percent['Defective Percent']-mean)/((std/count)**0.5)
p_value = []
p_value.append(norm.cdf(z_score.iloc[0]))
p_value.append(norm.cdf(z_score.iloc[1]))
p_value.append(norm.cdf(z_score.iloc[2]))
p_value.append(norm.cdf(z_score.iloc[3]))
p_value
```

```python
defective_percent['p_value']=p_value
defective_percent['similarity'] = defective_percent['p_value']>0.9
defective_percent
```

```python
defective_percent['similarity'] = defective_percent['similarity'].map({True:'similar',False:'disimilar'})
defective_percent
```
